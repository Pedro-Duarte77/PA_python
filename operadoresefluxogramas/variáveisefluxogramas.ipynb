## Cálculo de Média

Um fluxograma pode representar o processo de cálculo da média de duas notas, mostrando as etapas de entrada das notas, a operação de soma e a divisão por 2 para obter a média.

```python
# Recebe as notas das provas
nota1 = float(input("Digite a nota da P1: "))
nota2 = float(input("Digite a nota da P2: "))
# Calcula a média
media = (nota1 + nota2) / 2
# Mostra a média
print(f"A média é: {media}")
```

## Conversão de Moeda

Este exemplo demonstra como o fluxograma de conversão de moeda pode ser implementado em Python utilizando a entrada do usuário, a cotação do dólar e a operação de divisão.

```python
# Recebe a cotação do dólar e o valor em reais
cotacao_dolar = float(input("Digite a cotação do dólar: "))
valor_reais = float(input("Digite o valor em reais: "))
# Converte para dólares
valor_dolar = valor_reais / cotacao_dolar
# Mostra o resultado
print(f"O valor em dólares é: ${valor_dolar:.2f}")
```

## Cálculo de Comissão

Outro exemplo é o cálculo de comissão. O fluxograma é convertido em código Python, utilizando o preço unitário, a quantidade vendida e o cálculo da comissão.

```python
# Recebe o preço unitário e a quantidade vendida
preco_unitario = float(input("Digite o preço unitário da peça: "))
quantidade_vendida = int(input("Digite a quantidade vendida: "))
# Calcula o valor total da venda e a comissão
valor_total_venda = preco_unitario * quantidade_vendida
comissao = valor_total_venda * 0.05
# Mostra os resultados
print(f"Valor total da venda: R${valor_total_venda:.2f}")
print(f"Comissão do vendedor: R${comissao:.2f}")
```

## Lista de Exercícios -Devem fazer os fluxogramas e os programas

1) O custo ao consumidor de um carro novo é a soma do custo de fábrica com a percentagem do distribuidor e dos impostos (aplicados, primeiro os impostos sobre o custo de fábrica, e depois a percentagem do distribuidor sobre o resultado). Supondo que a percentagem do distribuidor seja de 28% e os impostos 45%. Escrever um algoritmo que leia o custo de fábrica de um carro e informe o custo ao consumidor do mesmo.

Exemplo: Custo de fábrica do carro: R$25.000,00

Impostos: 45% = R$11.250 (Total R$36.250,00)

Distribuição 28% = R$10.150 (Total R$ 46.400,00)

Para o consumidor: R$ 46.400,00

```python
# custo de fábrica
custo_fabrica = float(input("Digite o custo de fábrica do carro: R$"))
# porcentagens
impostos_percentual = 45 / 100
distribuidor_percentual = 28 / 100
# Calcula os impostos sobre o custo de fábrica
valor_impostos = custo_fabrica * impostos_percentual
preco_com_impostos = custo_fabrica + valor_impostos
# distribuição
distribuidor = preco_com_impostos * distribuidor_percentual
# custo final ao consumidor
custo_final = preco_com_impostos + distribuidor
# Resultado
print(f"O custo final do carro para o consumidor é: R${custo_final:.2f}")
```

3)  Continuando do exercício anterior, um funcionário trabalha em determinada empresa e quer saber quanto irá ganhar de hora extra. Você deverá fazer um algoritmo que solicite quantas horas e quantos minutos ele trabalhou, solicite horas e minutos de forma separada. Depois pergunte quanto ele ganha por hora e informe quanto ele irá ganhar. Sabendo que nessa empresa o funcionário ganha 70% a mais pelas horas trabalhadas.

Ex. Trabalhei neste mês 15h45min == 15.75

Ganho R$ 25 reais por hora

Resultado tem que ser = R$ 669,37 com os 70% amais

Outro teste de mesa: Se trabalhei 20h, devo ganhar R$ 850,00



```python
# Horas e minutos trabalhados
horas_trabalhadas = float(input("Informe as horas trabalhadas: "))
minutos_trabalhados = float(input("Informe os minutos trabalhados: "))
valor_por_hora = float(input("Informe quanto você ganha por hora: "))
# Conversão dos minutos para horas decimais
minutos_decimais = minutos_trabalhados / 60
# Total de horas trabalhadas em formato decimal
total_horas_trabalhadas = horas_trabalhadas + minutos_decimais
# Cálculo do salário normal
salario_normal = total_horas_trabalhadas * valor_por_hora
# Ganho com 70%
salario_extra = salario_normal * 1.70
# Resultado
print(f"Você irá ganhar {salario_extra}")
```

4) Simulação de Investimento em CDB com Juros Compostos

Uma pessoa deseja investir um valor em um CDB que rende uma taxa fixa de X% ao ano. Escreva um algoritmo que solicite o valor inicial do investimento, a taxa de juros anual (em %) e o período em anos. Depois, calcule o montante final considerando juros compostos.

 Fórmula dos juros compostos:

M=C×(1+i)**t (O t é potência)

onde:

M = montante final (valor após os anos de investimento)

C = capital inicial (valor investido)

i = taxa de juros anual (em decimal, ou seja, 5% = 0.05)

t = tempo em anos

 Exemplo:

Valor investido: R$ 1.000,00

Taxa de juros: 6% ao ano (0.06)

Tempo: 5 anos

```python
# Informações iniciais
valor_inicial = float(input("Digite o valor inicial do investimento: "))
taxa_juros_anual = float(input("Digite a taxa de juros anual: "))
periodo_anos = float(input("Digite o periodo em anos: "))
# Fórmula para descobrir o montante
montante = valor_inicial * (1+taxa_juros_anual/100)**periodo_anos
# Resultado
print(f"O montante será de {montante}")
```

## 5) Divisão de Conta em um Restaurante com Gorjeta

Um grupo de amigos foi a um restaurante e deseja dividir a conta igualmente entre eles. Além disso, o restaurante cobra uma taxa de serviço de 10% sobre o valor total da conta mais uma gorjeta.

Escreva um algoritmo que solicite o valor da conta e o número de pessoas no grupo. Depois, calcule o valor total com a taxa de serviço e quanto cada pessoa deve pagar.

 Fórmulas:

Definição da Gorjeta: Solicitar valor Gorjeta

Taxa de serviço: Aplicar 10% da taxa de serviço do restaurante, somar ao valor total e depois somar a gorjeta, já que o garçom foi muito solicito;

Valor por pessoa: Dividor pelo numero de pessoas

 Exemplo:

Valor da conta: R$ 200,00

Valor da Gorjeta: R$ 50,00

Número de pessoas: 4

Taxa de serviço (10% de R$ 200,00): R$ 20,00

Valor total a ser pago: R$ 220,00

Valor com a Gorjeta: R$270,00

Valor por pessoa: R$ 67,50

```python
# Solicita os valores de entrada
valor_conta = float(input("Digite o valor total da conta: "))
valor_gorjeta = float(input("Digite o valor da gorjeta: "))
num_pessoas = int(input("Digite o número de pessoas no grupo: "))

# Calcula a taxa de serviço (10% sobre o valor da conta)
taxa_servico = valor_conta * 0.10

# Calcula o valor total a ser pago (conta + taxa de serviço + gorjeta)
valor_total = valor_conta + taxa_servico + valor_gorjeta

# Calcula o valor por pessoa
valor_por_pessoa = valor_total / num_pessoas

# Resultados detalhados
print(f"\nValor total a ser pago: {valor_total:.2f}")
print(f"Valor por pessoa: {valor_por_pessoa:.2f}")
```
